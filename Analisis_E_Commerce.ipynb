```python
# Blok 1
import pandas as pd

# Membaca setiap file CSV ke dalam DataFrame
try:
    items_df = pd.read_csv('olist_order_items_dataset.csv')
    orders_df = pd.read_csv('olist_orders_dataset.csv')
    products_df = pd.read_csv('olist_products_dataset.csv')
    customers_df = pd.read_csv('olist_customers_dataset.csv')
    payments_df = pd.read_csv('olist_order_payments_dataset.csv')
    print("Semua file berhasil dimuat.")
except FileNotFoundError as e:
    print(f"Error: Pastikan file {e.filename} sudah di-upload ke Colab.")
```

```python
# Blok 2
# Menggabungkan tabel item dengan tabel pesanan
# 'inner' join berarti hanya baris yang punya pasangan order_id di kedua tabel yang akan diambil
merged_df = pd.merge(items_df, orders_df, on='order_id', how='inner')

print("Ukuran data setelah penggabungan pertama:", merged_df.shape)
merged_df.head()
```

```python
# Blok 5
# Menggabungkan dengan tabel produk
merged_df = pd.merge(merged_df, products_df, on='product_id', how='inner')

print("Ukuran data setelah penggabungan kedua:", merged_df.shape)
merged_df.head()
```

```python
# Blok 4
# Langsung gabungkan hasil dari Blok 3 dengan tabel pelanggan
# karena 'customer_id' sudah ada di merged_df
merged_df = pd.merge(merged_df, customers_df, on='customer_id', how='inner')

print("Ukuran data setelah penggabungan ketiga:", merged_df.shape)
merged_df.head()
```

```python
# Blok 5
# Terakhir, gabungkan dengan tabel pembayaran
final_merged_df = pd.merge(merged_df, payments_df, on='order_id', how='inner')

print("Ukuran data akhir:", final_merged_df.shape)
# Tampilkan beberapa kolom penting saja untuk verifikasi
final_merged_df[['order_id', 'product_category_name', 'customer_city', 'payment_type', 'price', 'payment_value']].head()
```

```python
final_merged_df.info()
```

```python
# Blok 6
# Cek jumlah data kosong di setiap kolom
print("Jumlah data kosong sebelum dibersihkan:")
final_merged_df.isnull().sum()
```

```python
# Blok 7
# Kolom yang akan diubah tipe datanya
kolom_tanggal = ['order_purchase_timestamp', 'order_approved_at', 'order_delivered_carrier_date', 'order_delivered_customer_date', 'order_estimated_delivery_date']

# Loop untuk mengubah setiap kolom tanggal menjadi tipe datetime
for kolom in kolom_tanggal:
    final_merged_df[kolom] = pd.to_datetime(final_merged_df[kolom], errors='coerce')

# Cek kembali tipe data untuk memastikan sudah berubah
final_merged_df.info()
```

```python
# Blok 8
# 1. Membuat kolom 'tahun' dan 'bulan' dari tanggal pembelian
final_merged_df['purchase_year'] = final_merged_df['order_purchase_timestamp'].dt.year
final_merged_df['purchase_month'] = final_merged_df['order_purchase_timestamp'].dt.month
final_merged_df['purchase_month_name'] = final_merged_df['order_purchase_timestamp'].dt.strftime('%B') # Nama bulan (e.g., January)

# 2. Menghitung durasi pengiriman dalam hari
final_merged_df['delivery_duration'] = (final_merged_df['order_delivered_customer_date'] - final_merged_df['order_purchase_timestamp']).dt.days

# Tampilkan beberapa kolom baru untuk verifikasi
final_merged_df[['order_purchase_timestamp', 'purchase_year', 'purchase_month_name', 'delivery_duration']].head()
```

```python
# Terakhir Metode Ekspor dengan Potongan (Chunking) dan Jeda Waktu

# Import library yang dibutuhkan
import pandas as pd
from google.colab import auth
from google.auth import default
import gspread
import time # library 'time' untuk memberi jeda
import numpy as np # Import numpy untuk menangani NaN

# Lakukan autentikasi
auth.authenticate_user()
creds, _ = default()
gc = gspread.authorize(creds)

# Buka spreadsheet dan pilih sheet pertama
nama_spreadsheet = "Olist E-commerce Data Mart"
spreadsheet = gc.open(nama_spreadsheet)
worksheet = spreadsheet.sheet1

# --- Logika Pengiriman Bertahap ---

# API Google Sheets tidak bisa menerima nilai seperti NaN atau inf.
# Kita harus mengubahnya menjadi format yang bisa diterima, seperti teks kosong ('').

final_merged_df = final_merged_df.replace([np.inf, -np.inf], np.nan) # Ubah nilai tak terhingga menjadi nilai NaN
final_merged_df = final_merged_df.fillna('') # Ubah semua nilai NaN menjadi teks kosong ('')

for col in final_merged_df.select_dtypes(include=['datetime64[ns]']).columns:
    final_merged_df[col] = final_merged_df[col].astype(str)

data_to_export = [final_merged_df.columns.values.tolist()] + final_merged_df.values.tolist()

# Tentukan ukuran potongan
chunk_size = 10000
total_rows = len(data_to_export)

print(f"Total baris yang akan dikirim: {total_rows-1}. Ukuran potongan: {chunk_size} baris.")
print("Memulai proses pengiriman... Ini akan memakan waktu beberapa menit.")

# Loop ini akan berjalan dari baris ke-0 hingga baris terakhir, dengan loncatan sebesar ukuran potongan
for i in range(0, total_rows, chunk_size):
    # Ambil potongan data
    chunk = data_to_export[i:i + chunk_size]

    # Kirim potongan ke worksheet
    worksheet.append_rows(chunk, value_input_option='USER_ENTERED')

    # Tampilkan progres
    # Adjust progress print to show the range of rows being sent
    print(f"Berhasil mengirim baris {i} hingga {min(i + chunk_size -1, total_rows-1)}...")


    # Beri jeda waktu 1.5 detik untuk "mengistirahatkan" API
    time.sleep(1.5)

print("\nSeluruh data berhasil diekspor ke Google Sheet.")
```
